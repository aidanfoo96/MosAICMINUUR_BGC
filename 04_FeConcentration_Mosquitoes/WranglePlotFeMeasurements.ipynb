```python
# Load packages 
import pandas as pd
```

```python
# read data 
Fe_data = pd.read_csv("120923_FeConc_LifeStages.txt", delimiter="\t")
Fe_data
```

```python
# function to split dataframes
def split_data(dataframe): 
    Positive_controls = dataframe[dataframe['group'].str.startswith("PC")]
    Experimental_groups = dataframe[~dataframe['group'].str.startswith("PC")]
            
    return Positive_controls, Experimental_groups

split_data(Fe_data)
```
